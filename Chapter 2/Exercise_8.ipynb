This exercise relates to the College data set, which can be found in the file College.csv on the book website. It contains a number of variables for 777 different universities and colleges in the US. 


The variables are

• Private : Public/private indicator

• Apps : Number of applications received

• P.Undergrad : Number of part-time undergraduates

• Outstate : Out-of-state tuition

• Room.Board : Room and board costs

• Books : Estimated book costs

• Personal : Estimated personal spending

• PhD : Percent of faculty with Ph.D.s

• Terminal : Percent of faculty with terminal degree

• S.F.Ratio : Student/faculty ratio

• perc.alumni : Percent of alumni who donate

• Expend : Instructional expenditure per student

• Grad.Rate : Graduation rate

# h)

Kao što je vidljivo na gornjim histogramima, većina izvrsnih studenata se nalazi na manjem broju fakseva, stoga, možemo pretpostaviti da su to nešto prestižniji fakulteti. Također je vidljivo da većina studenata cilja na isti, mali broj fakulteta, ti fakulteti bi se u toj sredini mogli smatrati prestižnima ili laganima.

Vidimo da postoji mali broj fakulteta koji troše relativno velike novce na svoje studente pa možemo istražiti kakvi su to fakulteti.

#

As can be seen in the histograms above, most excellent students are concentrated in a smaller number of universities, so we can assume these are somewhat more prestigious institutions. It is also evident that the majority of students aim for the same small group of universities, which in that region could be considered prestigious or easy to get into.

We can see that there are a few universities that spend relatively large amounts of money on their students, so we can explore what kind of institutions these are.

```python
fig, ax = plt.subplots(figsize=(5, 5))
college.boxplot('Expend', by='Private', ax=ax);
plt.ylabel('Expend')
```

Zaključujemo da je taj mali broj fakseva privatan što se moglo i pretpostaviti jer sigurno imaju više novčanih sredstava na raspolaganju.

Da ostanemo na sličnoj temi, pitajmo se, postoji li nekakva korelacija između novčane potrošnje fakulteta na studente i stope diplomiranih osoba na fakultetima.

#

We conclude that this small number of universities is private, which could have been expected since they likely have more financial resources available.

Staying on a similar topic, let's ask ourselves: is there any correlation between university spending per student and graduation rates at those universities?

```python
ax = college.plot.scatter('Expend', 'Grad.Rate')
ax.set_title('Expend vs. Grad.Rate');
```

Ako se fokusiramo na potrošnju po studentu ispod 20 000, na oko je teško vidjeti korelaciju. Međutim, zanimljivo je promotriti potrošnju iznad 30 000, gdje je vidljivo da je stopa diplomiranih jako visoka, a kako znamo da su to privatni fakulteti, to nam naslućuje da više ljudi završava privatne fakultete, tj. pretpostavili bismo da su lakši.

Naravno, za detaljniju analizu tog pitanja bilo bi pravilnije razdvojiti studente na privatne i one koji to nisu pa promatrati na dva različita grafa ili drugom bojom na istom.

Imamo naravno i jedan izuzetak, gdje malo ljudi diplomira na privatnom faksu koji puno troši na studente, što znači da je fakultet kvalitetan ili maksimalno zarađuje na studentima.

Također je zanimljiv slučaj fakulteta s Grad.Rate od 118, moguće da je tu došlo do neke greške pa bi s tim podatkom trebalo oprezno baratati.

Može se još puno igrati s danim podacima, međutim, tu ćemo stati.
