```python
column_names = list(Auto.columns.values)
fig, axs = plt.subplots(3, 2,figsize=(10, 10))
k = 0
for i in range (0,3):
    for j in range (0,2):
        k+=1
        Auto.plot.scatter(column_names[k], 'mpg', ax=axs[i, j]);

Auto.plot.scatter('origin', 'mpg',figsize=(5, 4));
```

Vidljiva je negativan korelacija između mpg i broja cilindara, mpg i displacement, mpg i konjskih snaga, mpg i težine, dok se pozitivna korelacija može vidjeti kod mpg i podrijetla i kod mpg i godine.

Negativna korelacija: mpg i broj cilindara, mpg i displacement, mpg i konjske snage, mpg i težine 

Pozitivna korelacija: mpg i podrijetlo, mpg i godine

Očito su nam najpouzdanije varijable za predviđanje mpg-a: displacement, broj konjskih snaga i težina jer je na tim grafovima vidljiva najmanja disperzija. Prirodno je pretpostaviti nastavak pozitivne korelacije između mpg-a i godine jer se očekuje utjecaj razvoja tehnologije, međutim, možemo očekivati i saturaciju.

#

A negative correlation is visible between mpg and the number of cylinders, mpg and displacement, mpg and horsepower, and mpg and weight, while a positive correlation can be seen between mpg and origin, and between mpg and year.

Negative correlation: mpg and number of cylinders, mpg and displacement, mpg and horsepower, mpg and weight

Positive correlation: mpg and origin, mpg and year

Clearly, the most reliable variables for predicting mpg are: displacement, horsepower, and weight, because these graphs show the least dispersion. It is natural to assume a continued positive correlation between mpg and year due to technological advancement, however, we can also expect saturation.
